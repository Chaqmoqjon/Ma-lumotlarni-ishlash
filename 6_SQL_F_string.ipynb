```python
import pandas as pd
import sqlite3 as sql
from pprint import pprint as print
```

```python
db = sql.connect("newdb.db") #yangi db yaratish
cursor =db.cursor()  #db ning cursori
```

```python
#yangi db yaratish shabloni
query = """
CREATE TABLE IF NOT EXISTS world_pop(
rank INT,
country TEXT PRIMERY KEY,
density DOUBLE,
population REAL NOT NULL,
area INT NOT NULL
);
"""
cursor.execute(query) # db ga kodlar ishlatildi
db.commit() #  #o'zgarishlarni faylga saqlash
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population.csv")
df.head()
```

```python
uz = df[df.country == "Uzbekistan"]
uz
```

```python
rank = uz.index.values[0]
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']

rank, country, density, population, area
```

```python
query = f"""
INSERT INTO world_pop VALUES({rank}, '{country}', {density}, {population}, {area})
"""
cursor.execute(query)  # f"" bilan ishlash kelajakda qiymatlarni funksiya orqali avtomatik berish imkonini beradi
db.commit()
```

```python
cursor.execute("SELECT * FROM world_pop")
cursor.fetchall()
```
