```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("newdb.db") #yangi db yaratish
cursor =db.cursor()  #db ning cursori
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/Electricity_Production_By_Source.csv")
df.head()
```

```python
cols = ['country', 'code', 'year', 'coal', 'gas', 'hydro', 'renewable', 'solar', 'oil', 'wind', 'nuclear']
df.columns = cols #ustun nomlari yangilandi
df.head()
```

```python
df.to_sql("electr_sql", db) #csv ni sql ga aylantirish, nomi va qaysi db ga joylangani (newdb ga joylandi)
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall() #db dagi datalar
```

```python
cursor.execute('PRAGMA table_info(electr_sql)')
cursor.fetchall()  #electr_sql ning info si
```

```python
query = "SELECT * FROM electr_sql WHERE country = 'Uzbekistan'"
df = pd.read_sql(query, db) #kerakli ma'lumotni olish
df.head()
```

##NEW DB yaratish

```python
#yangi db yaratish shabloni
query = """
CREATE TABLE IF NOT EXISTS world_pop(
rank INT,
country TEXT PRIMERY KEY,
density DOUBLE,
population REAL NOT NULL,
area INT NOT NULL
);
"""
cursor.execute(query) # db ga kodlar ishlatildi
db.commit() #  #o'zgarishlarni faylga saqlash
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall() # tablelarni ko'rish
```

```python
cursor.execute("PRAGMA table_info(world_pop);")
cursor.fetchall() #world_pop info
```

```python
query = "INSERT INTO world_pop VALUES (1, 'Macau', 21946.55, 658.256, 30)" #hamma qiymatlariga, qiymat berish uchun
cursor.execute(query)
db.commit()  #o'zgarishlarni faylga saqlash
```

```python
cursor.execute("SELECT * FROM world_pop") #ma'lumotni ko'rish
cursor.fetchall()
```

```python
query = "INSERT INTO world_pop (country, population, area) VALUES ('Uzbekistan', 33935.7, 447400)" #faqat berilgan qiymatlarga, qiymat beradi
cursor.execute(query) #kodni berish
db.commit() # saqlash
```

```python
cursor.execute("SELECT * FROM world_pop") #ma'lumotni ko'rish, ma'lumot berilmasa None bo'ladi
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM world_pop", db) #pandas bilan ko'rish (newdb dan kelayapti)
```

##UPDATE SQL

```python
query = "UPDATE world_pop SET rank = 168, density = 75.8511 WHERE country = 'Uzbekistan'" #update qilish
cursor.execute(query)
db.commit()  #o'zgarishlarni faylga saqlash
```

```python
pd.read_sql("SELECT * FROM world_pop", db)
```
