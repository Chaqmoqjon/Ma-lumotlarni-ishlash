```python
db  = sql.connect('chinook.db') #.db faylga ulanish
cursor = db.cursor() #cursor yaratish, sql kod yoza olish uchun
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table';") #db dagi barcha table nomlari
cursor.fetchall() #hammasini ekranga chiqarish
```

```python
cursor.execute("SELECT COUNT(*) FROM customers") #customerdagi qiymatlarni sanash
cursor.fetchone() #bitta qiymatni chiqarish
```

```python
cursor.execute("PRAGMA table_info(customers);") #customerdagi ustunlarning ma'lumotlari
cursor.fetchall()
```

```python
cursor.execute('SELECT * FROM customers') #hamma ma'lumotlarni olish
mijozlar = cursor.fetchall() #o'garuvchiga yuklash
```

```python
df = pd.read_sql('SELECT * FROM customers LIMIT 5', db, index_col='CustomerId') #sql code => database name => indeks column
df
```

```python
df = pd.read_sql('SELECT CustomerId, FirstName,	LastName,	City FROM customers', db, index_col='CustomerId') #sql code => database name => indeks column
df.head()
```

```python
komanda = 'SELECT CustomerId, FirstName,	LastName,	City FROM customers ORDER BY FirstName' # ismi bo'yicha tartiblash
df = pd.read_sql(komanda, db, index_col='CustomerId') #sql code => database name => indeks column
df.head()
```
