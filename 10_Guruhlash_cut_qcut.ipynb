```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population.csv', usecols = ['country', 'pop2021', 'area'],  index_col='country')
df.head()
```

```python
df.loc['Uzbekistan']
```

```python
df.loc['India']
```

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000] #guruhlahs uchun oraliqlar
```

```python
populatuion = df.pop2021
populatuion
```

```python
guruh = pd.cut(populatuion, bins) #guruhlash
guruh
```

```python
guruh.value_counts() #guruhlar bo'yicha qiymatlar
```

```python
df['guruh'] = guruh #df ga qo'shish
df
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv',  index_col=0)
df2.head()
```

```python
df2.price.describe() #guruhlash uchum min() va max() bilib olindi
```

```python
bins = [0, 10000, 20000, 50000]
names = ['arzon', 'o\'rta', 'qimmat']
```

```python
price_group = pd.cut(df2.price, bins, labels=names) #guruhlash va  nomini o'zgartirish
price_group.value_counts()
```

```python
df2['price_group'] = price_group
df2
```

```python
avto_range = pd.cut(df2.price, 4) #avtomatik guruhlash, qiymatlarni,  qiymatlar nisbati olinadi
avto_range.value_counts()
```

```python
avto_range = pd.qcut(df2.price, 4) #sanoq raqamini, nisbati olinadi
avto_range.value_counts()
```
