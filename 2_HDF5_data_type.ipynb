```python
import pandas as pd
```

##HDF5 formati - katta ma'lumotlarni saqlashda foydalaniladi. Ko'p dasturlash tillarida ham ichib ishlatsa bo'ladi

```python
df = pd.read_excel('/content/uzbekistan.xlsx', index_col=0, names= ['hudud', 'maydon', 'aholi']) #excel fayl uchun
df2 = pd.read_csv('/content/automobile_data.csv', index_col='index', nrows=10) #csv fayl uchun
```

```python
hdf_data = pd.HDFStore('data.h5') # HDF5 ga nom berib saqlandi
```

```python
hdf_data.put('uz_data', df) #uz_data nomi bilan saqlandi
```

```python
hdf_data.put('car_data', df2) #car_data nomi bilan saqlandi, bir nechta df larni saqlash mumkin
```

```python
hdf_data.keys() #ichidagi qiymatlarni ko'rish
```

```python
hdf_data['uz_data'] #dic kabi murojaat qilinaveradi
```

```python
hdf_data.close() #ma'lumotlar saqlanib bo'lgandan keyin berkitib qo'yish lozim
```
