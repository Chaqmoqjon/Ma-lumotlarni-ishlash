```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/MOCK_USER_DATA.csv', index_col='id')
df.head()
```

```python
df.describe() #noodatiy ma'lumotlarni tozlalash kerak
```

```python
df[(df.age > 123) | (df.age < 0)] # noodatiy qiymatlar ro'yxati
```

```python
df['age'] = df.age.apply(np.abs) #manfiy yoshlarni musbatga aylantirdik
```

```python
inx =  df[df.age > 123].index #o'chirish uchun indeksni olish
```

```python
df.drop( index = inx, inplace=True) #o'chirib yuborish
```

```python
df.describe() #noodatiy ma'lumotlar tozalandi
```
