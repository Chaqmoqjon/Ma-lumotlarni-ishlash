```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("electricity.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table'")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(elect)")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 10", db, index_col = "index")
```

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query, db)
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query = "SELECT * FROM elect WHERE country = 'Uzbekistan' AND year > 2000 "
uz = pd.read_sql(query, db)
uz.head()
```

```python
query = "SELECT country, year, coal, gas, hydro, oil  FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2005 and 2015 "
pd.read_sql(query, db) # 2005 va 2015 oralig'idagi ma'lumotlarni oling, BETWEEN
```

```python
query = "SELECT country, year, coal, gas, hydro, oil  FROM elect WHERE country = 'Uzbekistan' AND year IN (1990, 2002, 2006, 2010) "
pd.read_sql(query, db) # berilgan yillarni chiqarish, IN
```

```python
#LIKE

#  %var => Anvar, Sarvar
# _ar => car, bar, tar
# __ar => star, bear
# %ol% => xola, bolalar, lola

query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019, 2020, 2021) "
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year = 2020 "
pd.read_sql(query, db)
```
