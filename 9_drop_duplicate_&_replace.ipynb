```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv')
df2.head()
```

```python
df2['num-of-cylinders'].unique()
```

```python
raqam = {'two' : 2, 'three' : 3, 'four' : 4, 'five' : 5, 'six' : 6, 'eight' : 8, 'twelve' : 12} #kerakli key lar bilan dict yaratish
df2['cylinders'] = df2['num-of-cylinders'].map(raqam) #dict key larni moslab almashtirdi
df2
```

```python
kurs = 12975
df2['price_uz'] = df2['price'].map(lambda x: x * kurs ) #USD to UZS
df2
```

```python
df2.replace(raqam) #raqamlarni o'zgartirish, almashtirish
```

```python
df2.replace(['four','six'],[4,6])
# df.replace({'four':4, 'six':6})
```
