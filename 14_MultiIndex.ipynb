##Amaliy mashq. Havo va suv sifat darajasi

```python
df = pd.read_csv('https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/refs/heads/master/world_cities_air_water_quality.csv')
df
```

```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df
```

```python
df.Region = df.Region.map(lambda x: x.replace('"', ''))
df.Country = df.Country.map(lambda x: x.replace('"', ''))
```

```python
df['Country'][2] #oldida bo'sh joy bor ekan, tozalash kerak
```

```python
df.Country = df.Country.map(lambda x: x.strip())
df.Country[2] #tozalandi
```

```python
df[df['Country'] == 'Germany'] #kerakli ma'lumotni chiqarish
```

```python
df.to_csv('Update_Data', index = False) #havfsizlik uchun saqlandi
```

```python
df1 = pd.pivot_table(df, index = ['Country', 'City'], values = ['AirQuality', 'WaterPollution'])
df1  #MultiIndex
```

```python
df1.loc[['Germany', 'Uzbekistan', 'Tajikistan']] #kerakli davlarni olib solishtirish mumkin
```

```python
df2 = df1.copy() #yangi amallarni bajarish uchun
df2
```

```python
df2['AirPollution'] = 100 - df2['AirQuality'] #qiymatlarni teskari qiymatli ustunlar yaratildi
df2['WaterQuality'] = 100 - df2['WaterPollution']
df2
```

```python
df3 = pd.DataFrame(df2.values, index= df2.index,  columns = [['Air', 'Air', 'Water', 'Water'],['AirPollution', 'AirQuality', 'WaterPollution', 'WaterQuality']]) #ustun uchun ham multiindex yaratildi
df3
```

```python
df3.Air #kerakli kategoriyalarni topish osonlashdi
```

```python
df3.columns.names = ['Object', 'Info'] #ustun qatorlariga nom berildi
df3
```

```python
df3.swaplevel('Country', 'City') #indeks ustunlarini almashtirish
```
