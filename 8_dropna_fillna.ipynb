##NaN larni o'chirish

```python
df.dropna()
```

```python
len(df.dropna()) #NaN qiymat tashlab yuborilgandan keyingi qiymati
```

```python
df.dropna(axis = 1) #ustun bo'yicha o'chirish
```

```python
df.dropna(axis = 1, thresh=48) #agar ustunda 48 ta ma'lumot bo'lsa ustun qoladi aks holda o'chadi
```

```python
df2 = df.dropna(subset=['price', 'horsepower']) # faqat biz bergan ustunni NaN ga tekshirib o'chiradi, inplace = True joyida saqlash
df2.shape
```

##NaN larni to'ldirish

```python
df.head()
```

```python
df.fillna(0) # NaN qiymatlar o'rniga 0 qo'yildi
```

```python
df.fillna({'horsepower' : round(np.mean(df['horsepower']), 2), 'wheel-base' : 95 }) # berilgan ustunlarni to'ldirish
```

```python
df.fillna(method = 'ffill') # forward fill, o'zidan yuqoridagindan qiymatni olib to'ldiradi
```

```python
df.fillna(method = 'bfill') # back fill, o'zidan orqadagindan qiymatni olib to'ldiradi
```

```python
df[df.company == 'toyota']
```

```python
df[df.company == 'toyota'] = df[df.company == 'toyota'].fillna(method = 'bfill') # toyota ning NaN qiymatlari to'dirildi
```

```python
df[df.company == 'toyota']
```

```python
df.fillna(round(df.mean(numeric_only=True), 2)) #barcha ustunlar uchun o'rtachani joylash
```

```python
dfh = df[df['body-style'] == 'hatchback']
dfh.fillna(dfh.mean(numeric_only=True)) #body-style bo'yicha to'ldirish
```
