```python
df['datetime'] = pd.to_datetime(df['Date'], format = "%m/%d/%Y") #qaytadan datetime ga o'tkazildi
```

```python
df['D_Time'] = pd.to_datetime(df['Time'], format = "%H:%M:%S").dt.time #qaytadan datetime ga o'tkazildi, agar soatni olmoqchi bo'lsang dt.time ni o'chir
```
