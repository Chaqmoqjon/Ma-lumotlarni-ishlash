```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv',  index_col=0)
df.head()
```

```python
df.company.value_counts() # kompaniyalarning mashinalari sanaldi
```

```python
df.groupby('company').mean(numeric_only=True) #kategoriya bo'yicha o'rtachasi topildi
```

```python
df.groupby('company').price.max() #price ning maximumi topildi
```

```python
df.groupby(['company', 'body-style']).max() #ikki va undan ortiq ustunlarni bersa ham bo'ladi
```

```python
df.groupby('company').agg([ 'min', 'max']) # bir nechta funksiyalarni ishlatish
```

```python
df.groupby('company').agg({'price':['max', 'min', np.mean]})
```
