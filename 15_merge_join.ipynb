##Sodda misollar

```python
A = pd.DataFrame({
    'Transport' : ['poyezd', 'avtomobil', 'kema'],
    'Tezligi' : ['70', '180', '45']
})
A
```

```python
B = pd.DataFrame({
    'Transport' : ['poyezd', 'avtomobil', 'tramvay'],
    'Narxi' : ['98000', '10000', '65000']
})
B
```

```python
merged = pd.merge(A, B) #inner join. default inner join. ikkalasida borini chiqarish
merged
```

```python
merged_outer = pd.merge(A, B, how='outer')  #ikkalasida borini ham yo'g'ini ha chiqarish
merged_outer
```

```python
merged_left = pd.merge(A, B, how='left') #chap tomondagi to'liq chiqadi
merged_left
```

```python
mergerd_right = pd.merge(A, B, how='right') #o'ng tomondagi to'liq chiqadi
mergerd_right
```

```python
C = pd.DataFrame({
    'Transport_turi' : ['poyezd', 'avtomobil', 'kema'],
    'Tezligi' : ['70', '180', '45']
})
D = pd.DataFrame({
    'Transport' : ['poyezd', 'avtomobil', 'tramvay'],
    'Narxi' : ['98000', '10000', '65000']
})
```

```python
merg = pd.merge(C, D, left_on='Transport_turi', right_on='Transport') #ustunlarning nomlar har xil bo'lsa, ustunlarning nomlarini ko'rsatib qo'yish kerak
merg
```

```python
C = C.set_index('Transport_turi') #indeksi matn
C
```

```python
D #indeksi id
```

```python
merged1 = pd.merge(C, D, left_index=True, right_on='Transport') #indeksi va id orqali birlashtirish
merged1
```

```python
D = D.set_index('Transport') #transport indeks bo'ldi
D
```

```python
merged2 = pd.merge(C, D, left_index=True, right_index=True) #indeksi va indeks orqali birlashtirish
merged2
```
